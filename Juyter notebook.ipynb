```python
import ast
#  ast.literal_eval() use karenge. Yeh function ek safe alternative hai eval()
# ast.literal_eval? Taaki agar input malicious ho (e.g., os.system('rm -rf /')), 
# toh yeh dangerous commands execute na kare.
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
         L.append(i['name'])
    return L 
# String ko list of dictionaries mein convert kiya ast.literal_eval se.
# Loop chalakar har dictionary ka 'name' extract kiya aur ek nayi list mein daala.
# Final list ko return kiya.
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse).astype(str)
movies['crew'] = movies['crew'].apply(collapse).astype(str)
movies['genres'] = movies['genres'].apply(collapse).astype(str)
movies['keywords'] = movies['keywords'].apply(collapse).astype(str)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
['love','loving','loved']
```

```python
ps.stem('loving')
```

```python
new['tags'] = new['tags'].fillna('')  
new['tags'] = new['tags'].apply(lambda x: x.strip()) 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
cv.get_feature_names_out()
# now  the words aren't repeating because of stem function
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
# instead of calculating the eucledian distance{tip to vvfjovvhofvonck kkn dfjkvoudfbiotip distance} we will calculate the cosine distance{angle} as angle 
# is small the more the similarity due to high dimension failure is decrese in euleadian distance.
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = similarity[index]
    movies_list= sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movies_list:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Gandhi')
```
